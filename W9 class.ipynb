```python
# visualise data with scatter plot and NumPy

import numpy as np #imports numypy
import matplotlib.pyplot as plt #imports matplotlib

x = np.random.rand(50) #random numbers
y = np.random.rand(50)

plt.scatter(x, y) #scatter plot

plt.title("Scatter plot of random data points")
plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show() #shows graph
```

```python
# understand how to perform basic data analysis using numpy, such as calculating mean, median and standard deviation
#import numpy module
import numpy as np
#define a list of numbers
data = [10, 15, 8, 9, 14, 17, 6, 12]
#convert the list to a numpy array
data_array = np.array(data)
#calculate and print mean, median and standard deviation
mean = np.mean(data_array)
median = np.median(data_array)
std_dev = np.std(data_array)
print(f"Mean: {mean}")
print(f"Median: {median}")
print(f"Standard Deviation: {std_dev}")
```
