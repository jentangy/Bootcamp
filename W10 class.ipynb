```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load csv data
df = pd.read_csv('data.csv')
```

```python
print(df)
```

```python
#calculate the mean of temperatue
from numpy import average

mean_temp = df['Temperature'].mean()
print(f'Mean Temperature: {mean_temp}')

#calculate max and min of temperature
max_temp = df['Temperature'].max()
min_temp = df['Temperature'].min()
print(f'Max Temperature: {max_temp}')
print(f'Min Temperature: {min_temp}')

#standard deviation of temperature
std_temp = df['Temperature'].std()
print(f'Standard Deviation of Temperature: {std_temp}')
```

```python
#plot the temperature data over time
plt.figure(figsize=(12, 5))
plt.plot(df['Date'], df['Temperature'], marker='o', linestyle='-', color='b')
plt.title('Temperature Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.xticks(rotation=90) #rotate x-axis labels for better readability
plt.tight_layout #padding for better layout
plt.show()
```

```python
#histogram distribution of temperature
plt.figure(figsize=(10, 8))
plt.hist(df['Temperature'], bins=20, color='skyblue', edgecolor='blue')
plt.title('Temperature Distribution')
plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()  # padding for better layout
plt.show()
```

```python
#scatter plot of temperature vs humidity
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature'], df['Humidity'], color='cornflowerblue', alpha=0.6)
plt.title('Temperature vs Humidity')
plt.xlabel('Temperature (C)')
plt.ylabel('Humidity (%)')
plt.grid(True)
plt.tight_layout()  # padding for better layout
plt.show()
```

```python
#scatter plot of temperature and humidity over time
plt.figure(figsize=(12, 6))
plt.scatter(df['Date'], df['Temperature'], color='cornflowerblue', label='Temperature', alpha=0.6)
plt.scatter(df['Date'], df['Humidity'], color='violet', label='Humidity', alpha=0.6)
plt.title('Temperature and Humidity Over Time')
plt.xlabel('Date')
plt.ylabel('Values')
plt.xticks(rotation=90)  # rotate x-axis labels for better readability
plt.legend()
plt.grid(True)
plt.tight_layout()  # padding for better layout
plt.show()
```

```python
# Make subplot of temperature over time, temperature distribution, temperature vs humidity and temperature and humidity over time
fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].plot(df['Date'], df['Temperature'], marker='o', linestyle='-', color='b')
axs[0, 0].set_title('Temperature Over Time')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Temperature (C)')
axs[0, 0].tick_params(axis='x', rotation=90)  # rotate x-axis labels for better readability
axs[0, 1].hist(df['Temperature'], bins=20, color='skyblue', edgecolor='blue')
axs[0, 1].set_title('Temperature Distribution')
axs[0, 1].set_xlabel('Temperature (C)')
axs[0, 1].set_ylabel('Frequency')
axs[0, 1].grid(axis='y', alpha=0.75)
axs[1, 0].scatter(df['Temperature'], df['Humidity'], color='cornflowerblue', alpha=0.6)
axs[1, 0].set_title('Temperature vs Humidity')
axs[1, 0].set_xlabel('Temperature (C)')
axs[1, 0].set_ylabel('Humidity (%)')
axs[1, 1].scatter(df['Date'], df['Temperature'], color='cornflowerblue', label='Temperature', alpha=0.6)
axs[1, 1].scatter(df['Date'], df['Humidity'], color='violet', label='Humidity', alpha=0.6)
axs[1, 1].set_title('Temperature and Humidity Over Time')
axs[1, 1].set_xlabel('Date')
axs[1, 1].set_ylabel('Values')
axs[1, 1].tick_params(axis='x', rotation=90)  # rotate x-axis labels for better readability
axs[1, 1].legend()
axs[1, 1].grid(True)
plt.tight_layout()  # padding for better layout
plt.show()
```
